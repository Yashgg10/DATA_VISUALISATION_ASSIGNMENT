### **1. What is a parameter?**
A parameter is a measurable factor or characteristic that defines a system or sets the conditions of its operation. In machine learning, parameters are the internal variables learned by the model from the training data, such as weights and biases in a neural network.

---

### **2. What is correlation?**
Correlation measures the strength and direction of the linear relationship between two variables. It is expressed as a value between -1 and 1:
- \(1\): Perfect positive correlation.
- \(-1\): Perfect negative correlation.
- \(0\): No correlation.

---

### **3. What does negative correlation mean?**
Negative correlation means that as one variable increases, the other decreases. For example, the more time spent on leisure, the less time available for work.

### **4. Define Machine Learning. What are the main components in Machine Learning?**
**Machine Learning** is the process of developing algorithms that allow computers to learn and make decisions without being explicitly programmed.  
**Main Components:**
- **Data**: Input for training the model.
- **Model**: Mathematical representation of the problem.
- **Features**: Variables used for predictions.
- **Training**: The process of teaching the model.
- **Evaluation**: Assessing the model's performance.

---

### **5. How does loss value help in determining whether the model is good or not?**
The **loss value** quantifies how far the model's predictions are from the actual target values. A lower loss value generally indicates better performance. Continuous high loss may suggest an underfitting model.

### **6. What are continuous and categorical variables?**
- **Continuous Variables**: Numeric variables with an infinite range (e.g., height, temperature).
- **Categorical Variables**: Variables representing categories or groups (e.g., color, gender).

### **7. How do we handle categorical variables in Machine Learning? What are the common techniques?**
**Handling Techniques:**
1. **Label Encoding**: Assigns a unique number to each category.
2. **One-Hot Encoding**: Converts categories into binary columns.
3. **Ordinal Encoding**: Preserves the order in categorical data.

### **15. What is causation? Explain the difference between correlation and causation with an example.**
- **Causation**: When one event causes another.
- **Difference**: Correlation indicates a relationship; causation confirms one variable influences another.
**Example**: Ice cream sales and drowning incidents are correlated but not causal; both increase due to hot weather.
